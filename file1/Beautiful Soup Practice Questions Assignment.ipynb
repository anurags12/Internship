```python
def wikiheader(url):
    header=[]
    page=requests.get(url)
    soup=BeautifulSoup(page.content)
    for h in soup.find_all('span',class_="mw-headline"):
        header.append(h.text)
    df=pd.DataFrame({'Headers':header})
    return df
```

```python
def imbd100movies(url):
    page=requests.get(url)
    name=[]
    rating=[]
    year=[]
    soup=BeautifulSoup(page.content)
    for n in soup.find_all('h3',class_="lister-item-header"):
        name.append(re.sub('[()0-9.\n]+','',n.text))
    for r in soup.find_all('div',class_="ipl-rating-star small"):
        rating.append(re.sub('\n','',r.text))
    for y in soup.find_all('span',class_="lister-item-year text-muted unbold"):
        year.append(re.sub('[()]','',y.text))
    df=pd.DataFrame({'Name':name,'Ratings':rating,'Year_of_Release':year},index=range(1,101))
    return df
```

```python
url='https://www.imdb.com/list/ls091520106/'
imbd100movies(url)
```

```python
def delhirestaurents(url):
    name=[]
    cuisine=[]
    loc=[]
    rating=[]
    images=[]
    page=requests.get(url)
    soup=BeautifulSoup(page.content)
    for n in soup.find_all('div',class_="restnt-info cursor"):
        name.append(n.text)
    for c in soup.find_all('span',class_="double-line-ellipsis"):
        cuisine.append(re.sub('₹.*[|]','',c.text))
    for l in soup.find_all('div',class_="restnt-loc ellipsis"):
        loc.append(l.text)
    for r in soup.find_all('div',class_="restnt-rating rating-4"):
        rating.append(r.text)
    r1=soup.find('div',class_="restnt-rating rating-5")
    rating.insert(5,r1.text)
    rating.insert(12,r1.text)
    for i in soup.find_all('img',class_="no-img"):
        images.append(i['data-src'])
    df=pd.DataFrame({'Name':name,'Cuisine':cuisine,'Location':loc,'Rating':rating,'Images_url':images},index=range(1,22))
    return df
```

```python
url='https://www.dineout.co.in/delhi-restaurants/welcome-back'
delhirestaurents(url)
```

```python
def former_presidents_of_india(url):
    pname=[]
    nth=[]
    page=requests.get(url)
    soup=BeautifulSoup(page.content)
    for n in soup.find_all('div',class_="desc-sec"):
        pname.append(re.sub('\d+[A-Za-z ]+|\n','',n.text))
        o=re.search('\d+[A-Za-z ]+',n.text)
        nth.append(o.group())
    df=pd.DataFrame({'Name_of_Presidnet':pname,'In_Order_Desc':nth},index=range(1,15))
    return df
```

```python
url='https://www.presidentofindia.gov.in/former-presidents'
former_presidents_of_india(url)
```
