Q1-- Scrape  the  
details  of  
most  viewed  videos  on  
YouTube  
from  Wikipedia.  
Url 
= https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_v  You need to find following details: 
A) Rank
B) Name
C) Artist
D) Upload Date
E) Views
E) Views  E) Views   

```python
# searching for most  viewed  videos  on  YouTube from  Wikipedia
driver.get('https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos')
```

```python
#scraping details of most viewed videos on youtube from wikipedia
name=[]
artist=[]
upload=[]
views=[]

# Name of the video
try:
    name_tag=driver.find_elements(By.XPATH,'//table[@class="sortable wikitable sticky-header static-row-numbers sort-under col3center col4right jquery-tablesorter"]/tbody/tr/td[1]')
    for n in name_tag:
        name.append(re.sub('\[\d+\]|"','',n.text))

# Artist of the video
    art_tag=driver.find_elements(By.XPATH,'//table[@class="sortable wikitable sticky-header static-row-numbers sort-under col3center col4right jquery-tablesorter"]/tbody/tr/td[2]')
    for a in art_tag:
        artist.append(a.text)

# upload date of the video
    date_tag=driver.find_elements(By.XPATH,'//table[@class="sortable wikitable sticky-header static-row-numbers sort-under col3center col4right jquery-tablesorter"]/tbody/tr/td[4]')
    for d in date_tag:
        upload.append(d.text)

# Views on the video
    views_tag=driver.find_elements(By.XPATH,'//table[@class="sortable wikitable sticky-header static-row-numbers sort-under col3center col4right jquery-tablesorter"]/tbody/tr/td[3]')
    for v in views_tag:
        views.append(v.text)
        
except NoSuchElementException as err:
    print(err)

print(len(name),len(artist),len(upload),len(views))
```

```python
# creating dataframe
df=pd.DataFrame({'Rank':range(1,31),'Name':name,'Artist':artist,'Upload DATE':upload,'Views(Billions)':views})
df
```

You need to find following details:   
A) Series   
B) Place   
C) Date   
D) Time   
Note: - From bcci.tv home page you have reach to the international fixture page through code.

```python
# opening mentioned url
driver.get('https://www.bcci.tv/')
```

```python
# clicking on fixtures
fixture=driver.find_element(By.XPATH,'/html/body/header/div[3]/div[2]/ul/div[1]/a[2]')
fixture.click()
```

```python
#clicking on upcoming fixtures
upcoming=driver.find_element(By.XPATH,'//div[@class="tab-modify matchStatusFitler d-flex inactive"]/div[2]')
upcoming.click()
```

```python
# Searching for upcoming fixture urls elements
u_element=driver.find_elements(By.XPATH,'//div[@class="match-card-bottom"]/a')
print(len(u_element))
```

```python
# storing match urls
url=[]
for u in u_element:
    url.append(u.get_attribute('href'))

print(len(url))
```

```python
# match details
match=[]
series=[]
place=[]
date=[]
tym=[]
 # storing match details by iterating each urls
for u in url:
    driver.get(u)
    time.sleep(10)

    # fixture 
    try:
        m_tag=driver.find_element(By.XPATH,'//div[@class="widget matchDetails hideMobOnly showForSpecialCase"]/div/ul/li[2]/span[2]')
        match.append(m_tag.text)
    except NoSuchElementException:
        match.append('-')

    # Series details of fixture
    try:
        s_tag=driver.find_element(By.XPATH,'//div[@class="widget matchDetails hideMobOnly showForSpecialCase"]/div/ul/li[1]/span[2]')
        series.append(s_tag.text)
    except NoSuchElementException:
        series.append('-')

    # Place of the fixture
    try:
        p_tag=driver.find_element(By.XPATH,'//div[@class="widget matchDetails hideMobOnly showForSpecialCase"]/div/ul/li[3]/span[2]')
        place.append(p_tag.text)
    except NoSuchElementException:
        place.append('-')

    # date of fixture
    try:
        d_tag=driver.find_element(By.XPATH,'//div[@class="scoreBox-bottom"]/div[1]')
        date.append(d_tag.text)
    except NoSuchElementException:
        date.append('-')

    # time of fixture
    try:
        t_tag=driver.find_element(By.XPATH,'//div[@class="scoreBox-bottom"]/div[2]')
        tym.append(t_tag.text)
    except NoSuchElementException:
        tym.append('-')

print(len(match),len(series),len(place),len(date),len(tym))
```

```python
# Creating DataFrame
df=pd.DataFrame({'Match':match,'Series':series,'Place':place,'Date':date,'Time':tym})
df
```

You have to find following details: A) Rank   
B) State   
C) GSDP(21-22)- at current prices   
D) GSDP(22-23)- at current prices   
E) Share(21-22)   
F) GDP($ billion)   
Note: - From statisticstimes home page you have to reach to economy page through code.


```python
# searching for mentioned url
driver.get('https://statisticstimes.com/')
```

```python
# going on indian economy page
eco=driver.find_element(By.XPATH,'//div[@class="navbar"]/div[2]/div/a[3]')
driver.get(eco.get_attribute('href'))
```

```python
# GDP of indian states page
states=driver.find_element(By.XPATH,'//ul[@style="list-style-type:none;margin-left:20px;"]/li[1]/a')
driver.get(states.get_attribute('href'))
```

```python
# Scraping indian states gdp details
rank=[]
name=[]
g21_22=[]
g22_23=[]
share=[]
gdp=[]

# Rank of the state
r_tag=driver.find_elements(By.XPATH,'//div[@id="header"]/div[2]/div[5]/div[1]/div/table/tbody/tr/td[1]')
for r in r_tag:
    rank.append(r.text)

# Name of the State
st_tag=driver.find_elements(By.XPATH,'//div[@id="header"]/div[2]/div[5]/div[1]/div/table/tbody/tr/td[2]')
for st in st_tag:
    name.append(st.text)

# GSDP (Cr INR at Current prices) 21-22 of the state
g1_tag=driver.find_elements(By.XPATH,'//div[@id="header"]/div[2]/div[5]/div[1]/div/table/tbody/tr/td[5]')
for g in g1_tag:
    g21_22.append(g.text)

# GSDP (Cr INR at Current prices) 22-23 of the states
g2_tag=driver.find_elements(By.XPATH,'//div[@id="header"]/div[2]/div[5]/div[1]/div/table/tbody/tr/td[4]')
for p in g2_tag:
    g22_23.append(p.text)

# Share 21-22 of the state
sh_tag=driver.find_elements(By.XPATH,'//div[@id="header"]/div[2]/div[5]/div[1]/div/table/tbody/tr/td[6]')
for sh in sh_tag:
    share.append(sh.text)

#  GDP ($billion) of the state
gd_tag=driver.find_elements(By.XPATH,'//div[@id="header"]/div[2]/div[5]/div[1]/div/table/tbody/tr/td[7]')
for gd in gd_tag:
    gdp.append(gd.text)

print(len(rank),len(name),len(g21_22),len(g22_23),len(share),len(gdp))
```

```python
# Creating dataFrame
df=pd.DataFrame({'Rank':rank,
                'State':name,
                'GSDP (Cr INR at Current prices) 21-22':g21_22,
                'GSDP (Cr INR at Current prices) 22-23':g22_23,
                'Share(21-22)':share,
                'GDP($ billion)':gdp})
df
```

C) Contributors count   
D) Language used  
Note: - From the home page you have to click on the trending option from Explore menu through code.

```python
# Searching github
driver.get('https://github.com/')
```

```python
# clicking on open source
open=driver.find_element(By.XPATH,'//nav[@class="mt-0 px-3 px-lg-0 mb-3 mb-lg-0"]/ul/li[3]/button')
open.click()
```

```python
# clicking on trending repositoties
d=open.find_element(By.XPATH,'/html/body/div[1]/div[1]/header/div/div[2]/div/nav/ul/li[3]/div/div[3]/ul/li[2]/a')
d.click()
```

```python
# Searching for url elements
urr=driver.find_elements(By.XPATH,'//h2[@class="h3 lh-condensed"]/a')
# storing url
url=[]
for r in urr:
    url.append(r.get_attribute('href'))

print(len(url))
```

```python
# scraping mentioned details from repositories
n=[]
l=[]
co=[]
des=[]
for u in url:
    driver.get(u)
    time.sleep(5)

    # Name of the repository
    try:
        nam=driver.find_element(By.XPATH,'//strong[@class="mr-2 flex-self-stretch"]/a')
        n.append(nam.text)
    except NoSuchElementException:
        n.append('-')
        
    # Language used in repository
    try:
        lan=driver.find_element(By.XPATH,'//div[@class="BorderGrid about-margin"]/div[6]/div/ul')
        l.append(re.sub('\d.+%','',lan.text).replace('\n',','))
    except NoSuchElementException:
        l.append('-')
        
    #  Contributors count of the repository
    try:
        c=driver.find_element(By.XPATH,'//div[@class="BorderGrid about-margin"]/div[5]/div/h2/a/span')
        co.append(c.get_attribute('title'))
    except NoSuchElementException:
        co.append('-')
        
    # Repository Description
    try:
        d=driver.find_element(By.XPATH,'//div[@class="hide-sm hide-md"]/p')
        des.append(d.text)
    except NoSuchElementException:
        des.append('-')

print(len(n),len(l),len(co),len(des))
```

```python
# Creating dataframe
df=pd.DataFrame({'Repository title':n,'Repository description':des,'Contributors count':co,'Language used':l})
df
```

E) Ratings   
F) Votes

```python
# Searching mentioned url
driver.get('https://www.imdb.com/list/ls512407256/')
```

```python
# Scrolling down to store all 200 url elements
for _ in range(2000):
    driver.execute_script("window.scrollBy(0,200)")
    ur=driver.find_elements(By.XPATH,'//div[@class="ipc-title ipc-title--base ipc-title--title ipc-title-link-no-icon ipc-title--on-textPrimary sc-b189961a-9 iALATN dli-title"]/a')

print(len(ur))
```

```python
# storing url
url=[]
for u in ur: #iterating each url elements to find urls
    url.append(u.get_attribute('href')) # Storing urls

print(len(url))
```

```python
# Scaraping Mentioned Details of the tv shows
name=[]
genre=[]
span=[]
rt=[]
rating=[]
vote=[]
for r in url: # Iterating url list
    driver.get(r) # Searching each url to scrape data
    time.sleep(15) # Stopping webdriver for 5 sec to scrape data from each url

    # Name of the TV show
    try:
        name_tag=driver.find_element(By.XPATH,'//span[@class="hero__primary-text"]')
        name.append(name_tag.text)
    except NoSuchElementException:
        name.append('-')

    # Genre of the show
    try:
        genre_tag=driver.find_element(By.XPATH,'//div[@class="ipc-chip-list__scroller"]')
        genre.append(genre_tag.text.replace('\n',','))
    except NoSuchElementException:
        genre.append('-')


    # Year Span of show
    try:
        span_tag=driver.find_element(By.XPATH,'//ul[@class="ipc-inline-list ipc-inline-list--show-dividers sc-d8941411-2 cdJsTz baseAlt"]/li[2]/a')
        span.append(span_tag.text)
    except NoSuchElementException:
        span.append('-')

    # Run Time Of shows each episodes
    try:
        rt_tag=driver.find_element(By.XPATH,'//ul[@class="ipc-inline-list ipc-inline-list--show-dividers sc-d8941411-2 cdJsTz baseAlt"]/li[4]')
        rt.append(rt_tag.text)
    except NoSuchElementException:
        rt.append('-')

    # Rating of show
    try:
        rating_tag=driver.find_element(By.XPATH,'//div[@class="sc-3a4309f8-0 bjXIAP sc-b7c53eda-5 cxlubq"]/div/div[1]/a/span/div/div[2]/div[1]/span[1]')
        rating.append(rating_tag.text)
    except NoSuchElementException:
        rating.append('-')

    # Total Votes the show got
    try:
        vote_tag=driver.find_element(By.XPATH,'//div[@class="sc-3a4309f8-0 bjXIAP sc-b7c53eda-5 cxlubq"]/div/div[1]/a/span/div/div[2]/div[3]')
        vote.append(vote_tag.text)
    except NoSuchElementException:
        vote.append('-')

print(len(name),len(genre),len(span),len(rt),len(rating),len(vote))
```

```python
# Creating DataFrame
df=pd.DataFrame({'Name':name,'Year Span':span,'Genre':genre,'RunTime':rt,'Ratings':rating,'Votes':vote})
df
```

Q8 Details of Datasets from UCI machine learning repositories.   
Url = https://archive.ics.uci.edu/  You 
have to find the following details:   
A) Dataset name   
B) Data type   
C) Task   
D) Attribute type   
E) No of instances   
F) No of attribute G) Year   
Note: - from the home page you have to go to the Show All Dataset page through code.

```python
# Searching mentioned url
driver.get('https://archive.ics.uci.edu/')
```

```python
# Clicking on element to show all datasets
datasets=driver.find_element(By.XPATH,'//div[@class="flex flex-wrap justify-center gap-5"]/a[1]')
datasets.click()
```

```python
# Scraping dataset urls from 10 pages
url=[]
for i in range(0,10):
    ur=driver.find_elements(By.XPATH,'//h2[@class="truncate text-primary"]/a')
    for u in ur:
        url.append(u.get_attribute('href'))
        
    next=driver.find_element(By.XPATH,'//div[@class="btn-group"]/button[2]')
    next.click()
    time.sleep(3)

print(len(url))
```

```python
# Scraping datsets details
name=[]
task=[]
dtype=[]
no_inst=[]
no_attr=[]
attr_type=[]
year=[]

#iterating urls
for u in url:
    driver.get(u)
    time.sleep(10) # Wait of 10 seconds
    # name of datsaet
    try: 
        n=driver.find_element(By.XPATH,'//h1[@class="text-3xl font-semibold text-primary-content"]')
        name.append(n.text)
    except NoSuchElementException:
        name.append('-')
        
    # task of dataset
    try:
        t=driver.find_element(By.XPATH,'//div[@class="grid grid-cols-8 gap-4 md:grid-cols-12"]/div[3]/p')
        task.append(t.text)
    except NoSuchElementException:
        task.append('-')

    # Datatype of datasets
    try:
        dt=driver.find_element(By.XPATH,'//div[@class="grid grid-cols-8 gap-4 md:grid-cols-12"]/div[1]/p')
        dtype.append(dt.text)
    except NoSuchElementException:
        dtype.append('-')

    # No. of instances
    try:
        i=driver.find_element(By.XPATH,'//div[@class="grid grid-cols-8 gap-4 md:grid-cols-12"]/div[5]/p')
        no_inst.append(i.text)
    except NosuchElementException:
        no_inst.append('-')

    # No of attributes
    try:
        a=driver.find_element(By.XPATH,'//div[@class="grid grid-cols-8 gap-4 md:grid-cols-12"]/div[6]/p')
        no_attr.append(a.text)
    except NoSuchElementException:
        no_attr.append('-')

    # Attribute type
    try:
        at=driver.find_element(By.XPATH,'//div[@class="grid grid-cols-8 gap-4 md:grid-cols-12"]/div[4]/p')
        attr_type.append(at.text)
    except NoSuchElementException:
        attr_type.append('-')

   # Dataset Year
    try:
        y=driver.find_element(By.XPATH,'//h2[@class="text-sm text-primary-content"]')
        year.append(re.search('\d{4}',y.text).group())
    except NoSuchElementException:
        year.append('-')


print(len(name),len(task),len(dtype),len(no_inst),len(no_attr),len(attr_type),len(year))
```

```python
# Creating DataFrame
df=pd.DataFrame({'Dataset name':name,'Data Type':dtype,'Task':task,'Attribute Type':attr_type,'No of Instances':no_inst,'No of Attribute':no_attr,'Year':year})
df
```
